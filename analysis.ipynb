## Fractions

```python
import joblib

import matplotlib.pyplot as plt

path_lz4 = 'random_forest/fractions/'

n_results = [joblib.load(f"{path_lz4}{filename}.lz4") for filename in [f"model_frac_{int(frac*100)}_other_vs_natural" for frac in [0.1, 0.25, 0.5, 0.75]]]
c_results = [joblib.load(f"{path_lz4}{filename}.lz4") for filename in [f"model_frac_{int(frac*100)}_other_vs_cultivated" for frac in [0.1, 0.25, 0.5, 0.75]]]

n_results.append(joblib.load(f"random_forest/baseline/model_other_vs_natural.lz4"))
c_results.append(joblib.load(f"random_forest/baseline/model_other_vs_cultivated.lz4"))
```

```python
f1_scores = [d['f1_score'] for d in n_results]
recalls = [d['recall'] for d in n_results]
precisions = [d['precision'] for d in n_results]

labels = ['10%', '25%', '50%', '75%', '100%']

x = range(len(labels))

plt.figure(figsize=(8, 5))
plt.bar(x, f1_scores, width=0.2, label='F1 Score', align='center', color='#66bb66')
plt.bar([i + 0.2 for i in x], recalls, width=0.2, label='Recall', align='center', color='#339933')
plt.bar([i + 0.4 for i in x], precisions, width=0.2, label='Precision', align='center', color='#006600')
plt.xticks([i + 0.2 for i in x], labels)
plt.ylabel('Score')
plt.xlabel('Frações')
plt.title('Métricas para pastagem natural')
plt.legend()
plt.tight_layout()
plt.show()
```
